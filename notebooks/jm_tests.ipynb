```python
import pandas as pd
import os
import pickle
import glob
```

```python
%%time
in_path = "/home/bbscar/code/iShapeNoise/src/df_5000.csv" 
out_path = "/home/bbscar/code/iShapeNoise/src/" #Path to save the pickle files to
chunk_size = 400000 #size of chunks relies on your available memory
separator = ","

reader = pd.read_csv(in_path, sep=separator, chunksize=chunk_size, encoding = 'ISO-8859-1', low_memory=False)    

for i, chunk in enumerate(reader):
    out_file = out_path + "/data_{}.pkl".format(i+1)
    with open(out_file, "wb") as f:
        pickle.dump(chunk, f, pickle.HIGHEST_PROTOCOL)
        
pickle_path = "/home/bbscar/code/iShapeNoise/src/" #Same Path as out_path i.e. where the pickle files are

data_p_files=[]
for name in glob.glob(pickle_path + "/data_*.pkl"):
   data_p_files.append(name)

data = pd.DataFrame([])
for i in range(len(data_p_files)):
    data = data.append(pd.read_pickle(data_p_files[i]), ignore_index=True)
```

```python
print(f" ** Memory usage of the file - {sum(data.memory_usage()) * 0.000001} MB for {len(data.index)} Rows")
```
